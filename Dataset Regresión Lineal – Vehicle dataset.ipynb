```python
#Porcentaje de datos para entrenamiento (train)=75% , y para prueba (test) = 25%

#Se crean variables Xdatos y Ydatos para almacenar la información de entrenamiento

datos = df[['Ano','Precio_Venta','km_Recorridos','Motor Cilindraje','Potencia_Maxima','Asientos']]
Xdatos = datos.drop('Precio_Venta',axis=1)#Se suben los datos de train sin la columna Precio_Venta
Ydatos = datos['Precio_Venta']#Se suben los datos de train con los datos de la columna Precio_Venta

#division del conjunto de datos en dos bloques uno de entrenamiento y el otro de prueba (train and test)
Xdatos_train, Xdatos_test, Ydatos_train, Ydatos_test = train_test_split(Xdatos, Ydatos, test_size=0.25, random_state=0)
```

```python
# Graficar los datos y la linea de regresión
X = np.array(datos[['Ano']])
Y = datos['Precio_Venta'].values

#crear y ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X,Y)

# imprimir la ecuación de la recta
print(f"La ecuación de la recta es Y = {modelo_regresion_lineal.intercept_} + {modelo_regresion_lineal.coef_[0]}*X")

plt.scatter(X,Y, color = 'blue',label='Datos')
plt.plot(X,modelo.predict(X), color='red',label='Linea de Regresión')
plt.rcParams['figure.figsize']=(20,9)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Predicir el valor de Y usando el modelo
Y_pred = modelo.predict(X)

# Calcular Metricas
mse = mean_squared_error(Y,Y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(Y,Y_pred)
r2 = r2_score(Y,Y_pred)


# imprimir las metricas
print(f"Error cuadratico medio (MSE) : {mse}")
print()
print(f"Raiz del error cuadratico medio (RMSE) : {rmse}")
print()
print(f"Error absoluto medio (MAE) : {mae}")
print()
print(f"El coeficiente de determinación (R2) : {r2}")
```
