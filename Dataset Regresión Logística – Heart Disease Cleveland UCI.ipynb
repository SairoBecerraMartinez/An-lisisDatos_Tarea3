```python
#Conteo de datos atipicos
trestbps_maximo = 140
chol_maximo = 300
thalach_maximo = 170
oldpeak_maximo = 2
slope_maximo = 1

Total_Atipicos_trestbps=((data['trestbps']>trestbps_maximo)).sum()
print("El total de los datos atipicos en trestbps es de {}".format(Total_Atipicos_trestbps))

Total_Atipicos_chol=((data['chol']>chol_maximo)).sum()
print("El total de los datos atipicos en chol es de {}".format(Total_Atipicos_chol))

Total_Atipicos_thalach=((data['thalach']>thalach_maximo)).sum()
print("El total de los datos atipicos en thalach es de {}".format(Total_Atipicos_thalach))

Total_Atipicos_oldpeak=((data['oldpeak']>oldpeak_maximo)).sum()
print("El total de los datos atipicos en oldpeak es de {}".format(Total_Atipicos_oldpeak))

Total_Atipicos_slope=((data['slope']>slope_maximo)).sum()
print("El total de los datos atipicos en slope es de {}".format(Total_Atipicos_slope))
```

```python
#filtro datos atipicos
data=data[(data['trestbps']<trestbps_maximo)]
data=data[(data['chol']<chol_maximo)]
data=data[(data['thalach']<thalach_maximo)]
data=data[(data['oldpeak']<oldpeak_maximo)]
data=data[(data['slope']<slope_maximo)]
```

```python
#Validar si se removieron datos atipicos

Total_Atipicos_trestbps=((data['trestbps']>trestbps_maximo)).sum()
print("El total de los datos atipicos en trestbps es de {}".format(Total_Atipicos_trestbps))

Total_Atipicos_chol=((data['chol']>chol_maximo)).sum()
print("El total de los datos atipicos en chol es de {}".format(Total_Atipicos_chol))

Total_Atipicos_thalach=((data['thalach']>thalach_maximo)).sum()
print("El total de los datos atipicos en thalach es de {}".format(Total_Atipicos_thalach))

Total_Atipicos_oldpeak=((data['oldpeak']>oldpeak_maximo)).sum()
print("El total de los datos atipicos en oldpeak es de {}".format(Total_Atipicos_oldpeak))

Total_Atipicos_slope=((data['slope']>slope_maximo)).sum()
print("El total de los datos atipicos en slope es de {}".format(Total_Atipicos_slope))
```

```python
#Porcentaje de datos para entrenamiento (train)=75% , y para prueba (test) = 25%

#Se crean variables Xdatos y Ydatos para almacenar la información de entrenamiento

Xdata = data.drop('condition',axis=1)#Se suben los datos de train sin la columna Precio_Venta
Ydata = data['condition']#Se suben los datos de train con los datos de la columna Precio_Venta

#division del conjunto de datos en dos bloques uno de entrenamiento y el otro de prueba (train and test)
Xdata_train, Xdata_test, Ydata_train, Ydata_test = train_test_split(Xdata, Ydata, test_size=0.25, stratify = Ydata, random_state=0)
```

```python
print(Xdata_train)
```

```python
print(Ydata_train)
```

```python
#construir modelo de regresión Logistica
modelo_regresion_logistic = LogisticRegression(solver="liblinear")#Liblinear = libreria para resolver problemas lineales de clasificacion y regresion
modelo_regresion_logistic.fit(Xdata_train,Ydata_train)

#crear modelo de prediccion
Prediccion = modelo_regresion_logistic.predict(Xdata_test)#La funcion Predict predice los valores a partir del conjunto de datos de validación
```

```python
print(Ydata_test)
```

```python
print(Prediccion)
```

```python
#Desempeño del modelo de regresion
print(classification_report(Ydata_test, Prediccion))
```

```python
#matrix de confusion del modelo
cm = confusion_matrix(Ydata_test, Prediccion)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
```

```python
#curva de precision del modelo
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Ydata_test,Prediccion)
plt.show()
```
